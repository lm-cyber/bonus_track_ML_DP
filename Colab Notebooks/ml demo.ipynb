```python
# по вопросам писать в тг @dolboyoogurt

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('https://courses.openedu.ru/assets/courseware/v1/431a30c4a3a2b3f265f8cce1a809aa0f/asset-v1:ITMOUniversity+MLDATAN+spring_2023_ITMO_bac+type@asset+block/pulsar_stars_new.csv')
data = data.query('MIP >= 10 & MIP <= 100')
print(f'Число строк в полученной выборке: {len(data)}')
```

```python
data.head(0)
```

```python
print(f'Выборочное среднее для столбца MIP: {data.MIP.mean():0.3f}')
```

```python
print(f'Минимальное значение столбца MIP: {data.MIP.min():0.3f}')
```

```python
data = data.sort_values(by='MIP')

Y = np.array(list(data.TG))

X = data.drop(columns = 'TG')

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=33, stratify=Y)
print(f'Максимальное значение столбца STDC: {X_train.STDC.max():0.3f}')
```

```python
scaler = MinMaxScaler().fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train),columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test),columns=X_test.columns)



print(f'Выборочное среднее для столбца STDIP: {X_train.STDIP.mean():0.3f}')
```

```python
log = LogisticRegression().fit(X_train, y_train)
pred = log.predict(X_test)

k = 0
for i in range(len(pred)):
    if pred[i] == y_test[i] == 1:
        k +=1
print(f'True Positive = {k}')
```

```python
from sklearn.metrics import confusion_matrix
con = confusion_matrix(y_test,pred)
534-267
```

```python
print(f'f1 score = {f1_score(y_test, pred):0.3f}')
```

```python
neigh = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
pred = neigh.predict(X_test)

k = 0
for i in range(len(pred)):
    if pred[i] == y_test[i] == 1:
        k +=1
print(f'True Positive = {k}')
```

```python
confusion_matrix(y_test,pred)
```
