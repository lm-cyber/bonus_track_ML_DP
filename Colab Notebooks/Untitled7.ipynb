```python
df = pd.read_csv('gdrive/MyDrive/persons_pics_train/persons_pics_train.csv')
test = pd.read_csv('gdrive/MyDrive/persons_pics_reserved/persons_pics_reserved.csv')
```

```python
list(df.columns)
```

```python
names = list(df['label'].value_counts().index)
codes = [i for i in range(len(names)) ]
df['label'].replace(names,codes, inplace=True)
```

```python
X = df.drop('label',axis = 1)
Y = df['label']
data_for_X =  np.rint((X.to_numpy())*255.0).astype(int)
X[X.columns] = data_for_X
len(data_for_X)
```

```python
len(data_for_X[0])
```

```python
input_shape = (2914,)
num_classes = 12

# Define the fully connected neural network
model = Sequential([
    Flatten(input_shape=input_shape),
    Dense(512, activation='relu'),
    Dropout(0.2),
    Dense(256, activation='relu'),
    Dropout(0.2),
    Dense(128, activation='relu'),
    Dropout(0.2),
    Dense(num_classes, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Display model summary
model.summary()
X
```

```python
x_data = X.to_numpy()
x_data = x_data.astype('float32') / 255.0  # Normalize the images
y = Y.to_numpy()

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(x_data, y, test_size=0.2, random_state=42, stratify=y)

# Train the model
model.fit(X_train, y_train, epochs=200, validation_data=(X_val, y_val))
```
