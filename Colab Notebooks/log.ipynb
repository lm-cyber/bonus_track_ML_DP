```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
train = pd.read_csv('gdrive/MyDrive/candy-data.csv')
test = pd.read_csv('gdrive/MyDrive/candy-test.csv')
```

```python
train
```

```python
train_=train
```

```python
train = train_[(train['competitorname']!='Candy Corn') & (train['competitorname']!='Reeses stuffed with pieces') & (train['competitorname']!='Nerds')]
```

```python
x_train = train[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']]
```

```python
y_train =train['Y']
```

```python
model =LogisticRegression(random_state = 2019, solver = 'lbfgs').fit(x_train, y_train)
```

```python
confusion_matrix(y_train,model.predict(x_train))
```

```python
threshold = 0.5
y_pred = (model.predict_proba(x_train)[:, 1] > threshold).astype('float')
confusion_matrix(y_train, y_pred)
```

```python
x_test = test[['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus', 'sugarpercent', 'pricepercent']]
y_test =test['Y']
```

```python
threshold = 0.5
y_pred = (model.predict_proba(x_test)[:, 1] > threshold).astype('float')
confusion_matrix(y_test, y_pred)
```

```python
confusion_matrix(y_test,model.predict(x_test))
```

```python
from sklearn.metrics import recall_score
from sklearn.metrics import auc
from sklearn.metrics import precision_score
from sklearn import metrics
```

```python

fpr, tpr, thresholds = metrics.roc_curve(y_test, model.predict(x_test))
# metrics.auc(fpr, tpr)
fpr
```

```python
# model.predict(x_test) == y_pred
recall_score(y_test,model.predict(x_test))
```

```python
precision_score(y_test,model.predict(x_test))
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(model.predict(x_test),y_test)
```
