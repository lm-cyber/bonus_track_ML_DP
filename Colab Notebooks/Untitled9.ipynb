```python
import pandas as pd 
import numpy as np
import catboost as cb
```

```python
df = pd.read_csv('https://courses.openedu.ru/assets/courseware/v1/0c61cf43490f9189f8ddb65a887dcc64/asset-v1:ITMOUniversity+MLDATAN+spring_2023_ITMO_bac+type@asset+block/Bank_Personal_Loan_Modelling_train.csv')
```

```python
df.head()
```

```python
df.drop(['ID','ZIP Code'],axis=1,inplace=True)
```

```python
from catboost import CatBoostClassifier
X=df.drop('Personal Loan',axis=1)
Y=df['Personal Loan']
from sklearn.model_selection import train_test_split
x,xv,y,yv = train_test_split(X,Y,stratify=Y)

model = CatBoostClassifier(
    iterations=100000,
    learning_rate=0.001,
    loss_function='Logloss',
    eval_metric='Accuracy',
    early_stopping_rounds=2000,
    
    # weight=w
        # stratify=True


    # task_type="GPU"
)

model.fit(
    X, Y,
    cat_features=['Education','Securities Account','CD Account','Online','CreditCard'],
    verbose=False,
    # plot=True,

)
```

```python
test =pd.read_csv('https://courses.openedu.ru/assets/courseware/v1/dddc0dd9f40894c81c3d3685d3a74f9d/asset-v1:ITMOUniversity+MLDATAN+spring_2023_ITMO_bac+type@asset+block/Bank_Personal_Loan_Modelling_reserved.csv')
test.drop(['ID','ZIP Code'],axis=1,inplace=True)
```
