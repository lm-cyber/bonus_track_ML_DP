```python
im_dir = sorted(list(paths.list_images('gdrive/MyDrive/train')))
```

```python
def e_hist(image, bins=(8, 8, 8)):
    hist = cv2.calcHist([image], [0, 1, 2], None, bins, [0, 256, 0, 256, 0, 256])
    cv2.normalize(hist, hist)
    return hist.flatten()
```

```python
model = LinearSVC(C=0.94, random_state=2)
data = [e_hist(cv2.imread(i)) for i in im_dir]
y = []
for i in im_dir:
    if 'cat' in i:
        y.append(1)
    else:
        y.append(0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.25, random_state=2)
```

```python
model = model.fit(X_train,y_train)
f1_score(model.predict(X_test),y_test, average='micro')
```
