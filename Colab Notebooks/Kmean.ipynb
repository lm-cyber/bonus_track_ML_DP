```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
```

```python
first = pd.read_csv('gdrive/MyDrive/kmean.csv')
```

```python
model = KMeans(n_clusters=3, init=np.array([[8.5, 8.75], [10.5, 7.13], [12.0, 5.0]]), max_iter=100, n_init=1)
```

```python
first
```

```python
X =first[['X','Y']]
```

```python
model = model.fit(X)
```

```python
test =model.labels_
```

```python
model.get_params()
```

```python
model.score(X)
```
